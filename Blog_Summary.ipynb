## 1. Collect list of newspaper articles

```python
from collections import defaultdict
from newspaper import Article

#A new article from TOI

url_dict = defaultdict(list)

url_dict = {'meta':     ["https://ai.facebook.com/blog/dino-self-supervised-learning-demo/", # 2 *
                         "https://ai.facebook.com/blog/generating-wikipedia-biographies/",
                         "https://ai.facebook.com/blog/how-ai-is-helping-address-the-climate-crisis/",],
            
            'deepmind':  ['https://www.deepmind.com/blog/predicting-the-past-with-ithaca', # 1
                          'https://www.deepmind.com/blog/accelerating-fusion-science-through-learned-plasma-control',
                          'https://www.deepmind.com/blog/muzeros-first-step-from-research-into-the-real-world',],
            
            'openai'  :  ['https://openai.com/blog/measuring-goodharts-law/', # 5
                          'https://openai.com/dall-e-2/',],
            
            'ms'      :  ['https://blogs.microsoft.com/ai/new-z-code-mixture-of-experts-models-improve-quality-efficiency-in-translator-and-azure-ai/',
                          'https://blogs.microsoft.com/ai/new-azure-openai-service/'], # 3 *
            
            'nvida'   :  ['https://blogs.nvidia.com/blog/2022/04/21/tooth-tech-ai-takes-bite-out-of-dental-slide-misses-by-assisting-doctors/',
                          'https://blogs.nvidia.com/blog/2022/04/19/in-the-nvidia-studio/',] # 4
            }
```

## 2. Print summary of articles by company

```python
company_list = list(url_dict.keys())

for company in company_list:
  print("#"*50)
  print("Company name: ", company)
  print("#"*50)
  print("")

  url_list = url_dict[company]
  for url in url_list:
    print("*"*200)

    #For different language newspaper refer above table
    toi_article = Article(url, language="en") # en for English
    #To download the article
    toi_article.download()
    #To parse the article
    toi_article.parse()
    #To perform natural language processing ie..nlp
    toi_article.nlp()

    #To extract title
    print("Article's Title:")
    print(toi_article.title)
    print("")

    #To extract summary
    print("Article's Summary:")
    print(toi_article.summary)
    print("")

    #To extract keywords
    print("Article's Keywords:")
    print(toi_article.keywords)

    print("*"*200)
    print("")
```
